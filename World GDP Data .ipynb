Importing required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

Loading the dataset and converting into a pandas dataframe from csv format

```python
df = pd.read_csv('D:\Cleaning Projects files\GDP Data\gdp data(main file).csv')
```

Looking at the first few rows of the dataset to make sure that the the dataset has loaded properly

```python
pd.set_option('display.max_columns', None)
df.head()
```

The reason we need to replace NaNs by zero is that most of the countries in the index weren't formed by that particular year and of the remaining few there is no data about the value of the GDP avilable on the internet.

```python
df.replace(np.NaN, 0, inplace = True)
```

Looking at the datatypes of the columns in the dataset

```python
df.dtypes
```

Let's build a basic line plot showing the transition of World GDP from 1960 to 2017

```python
# Creating a line plot of global GDP  from 1960-2017

import matplotlib.pyplot as plt
%matplotlib inline

#Setting the country name as index and applying the change to the dataframe by setting the inplace parameter to True
df.set_index('Country Name', inplace = True)

#Creating a list of years nesting the range function in the map function
yr = list(map(str,range(1960,2017,5)))

#Accessing the newly  created and the plot method to plot the line chart by setting the kind parameter to line
df.loc['World', yr].plot(kind = 'line', figsize = (10,6))

# Details of the line plot
plt.xlabel('year')
plt.ylabel('GDP(in millions USD)')
plt.title('World GDP(1960-2017)')
plt.show()
```

Let's do the same for a single country (say Afghanistan)

```python
yr = list(map(str,range(1960,2017,2)))
df.loc['Afghanistan', yr].plot(kind = 'line', figsize = (10,6))
plt.xlabel('year')
plt.ylabel('GDP(in millions USD)')
plt.title('Afghanistan GDP 1960-2017')
plt.show()
```
